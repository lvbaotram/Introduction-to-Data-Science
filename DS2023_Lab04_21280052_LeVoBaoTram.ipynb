## 1. Web Scraping:

### Import libraries

```python
from bs4 import BeautifulSoup
import urllib.request
import requests
import numpy as np
from random import randint
import pandas as pd
import re
```

```python
def download_html(url):
    with urllib.request.urlopen(url) as response:
        html = response.read()
        html = html.decode('utf-8')
    response.close()
    return html
```

### Random page on website with by 25

```python
pages = np.arange(0,718,25)
```

```python
pages
```

### Retrieve the links to the articles
- Access the information of "Abstract", "Download URL"

```python
abstract_data=[]
download_data=[]
for page in pages:
  page =requests.get('https://arxiv.org/list/cs.AI/pastweek?skip='+str(page)+'&show=25')
  soup = BeautifulSoup(page.text, 'html.parser')
  data = soup.find_all("a", title ="Abstract", href = True)
  for i in data:
    abstract= "https://arxiv.org" +i.get('href')
    abstract_data.append(abstract)
  data_pdf = soup.findAll('span',{'class':"list-identifier"})
  for i in data_pdf:
    i = soup.find("a", title ="Download PDF", href = True)
    if i is None:
      download = 'None';
    if i is not None:
       download= "https://arxiv.org" +i.get('href')
    download_data.append(download)
```

```python
#The link to the articles
abstract_data
```

- Next, access the information: "Title", "Author", "Subject",

```python
title_data=[]
author_data=[]
subject_data=[]
for page in pages:
  page =requests.get('https://arxiv.org/list/cs.AI/pastweek?skip='+str(page)+'&show=25')
  soup = BeautifulSoup(page.text, 'html.parser')
  data = soup.findAll('div',  {'class':'meta'})
  j = 0
  for i in data:
    title = i.find('div', attrs ={"class":"list-title mathjax"}).text.replace("\n", "").replace("Title:","")
    title_data.append(title)
    author = i.find('div', attrs = {"class":"list-authors"}).text.replace("\n","").replace("Authors:","")
    author_data.append(author)
    subject = i.find("div",attrs = {"class":"list-subjects"}).text.replace("\n","").replace("Subjects:","")
    subject_data.append(subject)
```

```python
df = pd.DataFrame({"Title" :title_data, "Author": author_data, "Abstract":  abstract_data , "Subject": subject_data, "Download URL":  download_data})
df
```

```python
df.to_csv('DS2023_Lab04_21280052_LeVoBaoTram',index=False)
```

## 2. Regular Expression

### Output all date formats present in the text.




```python
with open("5.txt","r",encoding="utf-8") as file:
    data = file.read()
    data= data.strip() 
```

```python
date= 'ngày \d{1,2} tháng \d{1,2} năm \d{4}'
re.findall(date,data)
```

### Split the text into a list of laws.




```python
law = r"(?s)(Điều \d+\..*?(?=Chương \d+|Điều \d+\.))"
laws = re.findall(law,data)
```

```python
for i, law in enumerate(laws, start=1):
    print(f"{law}")
```

### Find all law codes present in the text using the format similar to: '204/2004/NĐ-CP'"

```python
code= re.findall(r"\d+/\d+/[\w\d-]+", data)
print(code)
```
